```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
import numpy as np

# Load the JSON file
with open("LSTM_loops.json", "r") as file:
    lstm_loops = json.load(file)

with open("CNN_loops.json", "r") as file:
    cnn_loops = json.load(file)

with open("GRU_loops.json", "r") as file:
    gru_loops = json.load(file)
    
# Convert JSON data to DataFrame
lstm_loops = pd.DataFrame([(error_rate, config, *metrics.values()) for error_rate, configs in lstm_loops.items() for config, metrics in configs.items()],
                  columns=['error_rate', 'config', 'accuracy_before', 'classification_error_before', 'accuracy_after', 'classification_error_after', 'classification_error_induced'])

lstm_loops['error_rate'] = pd.to_numeric(lstm_loops['error_rate'])

cnn_loops = pd.DataFrame([(error_rate, config, *metrics.values()) for error_rate, configs in cnn_loops.items() for config, metrics in configs.items()],
                  columns=['error_rate', 'config', 'accuracy_before', 'classification_error_before', 'accuracy_after', 'classification_error_after', 'classification_error_induced'])

cnn_loops['error_rate'] = pd.to_numeric(cnn_loops['error_rate'])

gru_loops = pd.DataFrame([(error_rate, config, *metrics.values()) for error_rate, configs in gru_loops.items() for config, metrics in configs.items()],
                  columns=['error_rate', 'config', 'accuracy_before', 'classification_error_before', 'accuracy_after', 'classification_error_after', 'classification_error_induced'])

gru_loops['error_rate'] = pd.to_numeric(gru_loops['error_rate'])
```

```python
combined_df = pd.concat([
    lstm_loops.assign(Model='LSTM'),
    cnn_loops.assign(Model='CNN'),
    gru_loops.assign(Model='GRU')
])

# Create a scatter plot
plt.figure(figsize=(10, 7))
sns.scatterplot(data=combined_df, x='error_rate', y='classification_error_induced', hue='Model', style='Model')

# Set log scale for x-axis
plt.xscale('log')

# Set the labels and title
plt.xlabel('Error Rate (log scale)')
plt.ylabel('Classification Error Induced')
plt.title('Fault Tolerance Comparison of Models')

# Display the plot
plt.grid(True)
plt.show()

combined_df = pd.concat([
    lstm_loops.assign(Model='LSTM'),
    cnn_loops.assign(Model='CNN'),
    gru_loops.assign(Model='GRU')
])

# Create a scatter plot
plt.figure(figsize=(10, 7))
sns.scatterplot(data=combined_df, x='error_rate', y='accuracy_before', hue='Model', style='Model')

# Set log scale for x-axis
plt.xscale('log')

# Set the labels and title
plt.xlabel('Error Rate (log scale)')
plt.ylabel('Classification Error Induced')
plt.title('Fault Tolerance Comparison of Models')

# Display the plot
plt.grid(True)
plt.show()

# Create a scatter plot
plt.figure(figsize=(10, 7))
sns.scatterplot(data=combined_df, x='error_rate', y='accuracy_after', hue='Model', style='Model')

# Set log scale for x-axis
plt.xscale('log')

# Set the labels and title
plt.xlabel('Error Rate (log scale)')
plt.ylabel('Classification Error Induced')
plt.title('Fault Tolerance Comparison of Models')

# Display the plot
plt.grid(True)
plt.show()
```

```python
from scipy.interpolate import interp1d
import statsmodels.api as sm

def plot_lowess(df, hue_col, x_col, y_col, title):
    unique_hues = df[hue_col].unique()

    plt.figure(figsize=(10, 7))

    # For each hue value, fit a LOWESS and plot
    for hue in unique_hues:
        subset = df[df[hue_col] == hue].sort_values(by=x_col)
        lowess = sm.nonparametric.lowess(subset[y_col], subset[x_col], frac=0.3)

        plt.plot(lowess[:, 0], lowess[:, 1], label=hue)

    # Set log scale for x-axis
    plt.xscale('log')

    # Set the labels and title
    plt.xlabel('Error Rate (log scale)')
    plt.ylabel('Classification Error Induced')
    plt.title(title)

    # Add a legend
    plt.legend()

    # Display the plot
    plt.grid(True)
    plt.show()
    
plot_lowess(
    df=combined_df, 
    hue_col='Model', 
    x_col='error_rate', 
    y_col='classification_error_induced', 
    title='Fault Tolerance Comparison of Models'
)

plot_lowess(
    df=combined_df, 
    hue_col='Model', 
    x_col='error_rate', 
    y_col='accuracy_before', 
    title='Fault Tolerance Comparison of Models Before'
)

plot_lowess(
    df=combined_df, 
    hue_col='Model', 
    x_col='error_rate', 
    y_col='accuracy_after', 
    title='Fault Tolerance Comparison of Models After'
)

# Plot for models
plot_lowess(combined_df, 'Model', 'error_rate', 'classification_error_induced', 'Fault Tolerance Comparison of Models (LOWESS)')

# Plot for LSTM configurations
plot_lowess(lstm_loops_relevant, 'config', 'error_rate', 'classification_error_induced', 'Fault Tolerance Comparison of LSTM Configurations (LOWESS)')

plot_lowess(gru_loops_relevant, 'config', 'error_rate', 'classification_error_induced', 'Fault Tolerance Comparison of GRU Configurations (LOWESS)')

plot_lowess(cnn_loops_relevant, 'config', 'error_rate', 'classification_error_induced', 'Fault Tolerance Comparison of CNN Configurations (LOWESS)')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
import numpy as np

# Load the JSON files into dictionaries
cnn_layers = json.load(open('CNN_layers.json'))
lstm_layers = json.load(open('LSTM_layers.json'))
gru_layers = json.load(open('GRU_layers.json'))

# Combine the dictionaries into a list for easy iteration
layers_list = [cnn_layers, lstm_layers, gru_layers]
layers_names = ['CNN', 'LSTM', 'GRU']
```

```python
df_list = []
for i in range(3):
    df = pd.json_normalize(layers_list[i])
    df['DNN'] = layers_names[i]
    df_list.append(df)

# Concatenate the dataframes into one
df = pd.concat(df_list)
```

```python
# Extract the necessary information from the DataFrame
df_melted = df.melt(id_vars=['DNN'], var_name='Layer_Config', value_name='Classification_Error_Induced')

# Split the 'Layer_Config' column into 'Layer' and 'Config' columns
df_melted[['Layer', 'Config']] = df_melted['Layer_Config'].str.split('.', 1, expand=True)

# Filter the DataFrame to only include rows where 'Config' ends with 'classification_error_induced'
df_melted = df_melted[df_melted['Config'].str.endswith('classification_error_induced')]

# Remove the 'classification_error_induced' part from the 'Config' column
df_melted['Config'] = df_melted['Config'].str.replace('.classification_error_induced', '')
```

```python
# Filter the DataFrame to only include rows for the 'signed-dense-2-6' configuration
df_filtered = df_melted[df_melted['Config'].str.contains('signed-dense-2-6')]

# Create a figure with three subplots, one for each DNN
fig, axs = plt.subplots(3, 1, figsize=(15, 15))

# Loop over the DNNs and create a barplot for each one
for i, dnn in enumerate(df_filtered['DNN'].unique()):
    # Filter the DataFrame to only include rows for the current DNN
    df_dnn = df_filtered[df_filtered['DNN'] == dnn]
    
    # Create a barplot
    sns.barplot(data=df_dnn, x='Layer', y='Classification_Error_Induced', color='blue', ax=axs[i])
    
    # Set the title of the subplot
    axs[i].set_title(f'{dnn} Layer-specific Fault-injection Analysis (signed-dense-2-6)')
    
    # Rotate the x-axis labels for readability
    axs[i].tick_params(axis='x', rotation=90)

# Adjust the layout of the subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import json
import zipfile


# File paths
file_paths = ['CNN_settings.json', 'LSTM_settings.json', 'GRU_settings.json']

# Loading JSON files into pandas dataframes
df_list = []
for file in file_paths:
    with open(file, 'r') as f:
        data = json.load(f)
    df = pd.json_normalize(data)
    df['Model'] = file.split('/')[-1].split('_')[0] # Extracting model name from file name
    df_list.append(df)

# Concatenating all dataframes
combined_df = pd.concat(df_list, ignore_index=True)

# Reshaping the DataFrame to have settings, model, and classification error induced as separate columns
reshaped_df = pd.melt(combined_df, id_vars=['Model'], 
                      value_vars=[col for col in combined_df.columns if 'classification_error_induced' in col], 
                      var_name='Setting', 
                      value_name='Classification_Error_Induced')

# Removing the '.classification_error_induced' part from the 'Setting' column for clarity
reshaped_df['Setting'] = reshaped_df['Setting'].str.replace('.classification_error_induced', '')

import seaborn as sns
import matplotlib.pyplot as plt

# Create a separate bar plot for each model
for model in reshaped_df['Model'].unique():
    plt.figure(figsize=(12,6))
    model_data = reshaped_df[reshaped_df['Model'] == model]
    sns.barplot(x='Setting', y='Classification_Error_Induced', data=model_data)
    plt.title(f'{model} Model: Setting vs. Classification Error Induced')
    plt.xticks(rotation=90)
    plt.show()
```
